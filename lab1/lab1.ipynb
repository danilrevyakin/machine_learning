```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_data(file_name):
    return np.genfromtxt(file_name, delimiter=",", skip_header=1)


def plot_data(data, title, color):
    x, y = list(data[:, 1]), list(data[:, 2])
    plt.title(title)
    plt.plot(x, y, color)
    

def linear_regression_function(beta0, beta1, x):
    return beta0 + np.multiply(beta1, x)


def loss_function(predicted_row, actual_row):
    result = sum((predicted_row[i] - actual_row[i]) ** 2 for i in range(len(predicted_row)))
    return result / len(predicted_row)
```

```python
train_data = get_data("lab_1_train.csv");
plot_data(train_data, "Train data", 'yo')
```

```python
tolerance = 0.000001
lr = 0.005


def train(rows):
    beta0, beta1 = np.random.randn(1), np.random.randn(1)
    x_column, y_column = list(rows[:, 1]), list(rows[:, 2])
    predicted_y = [linear_regression_function(beta0, beta1, x) for x in x_column]
    loss = loss_function(predicted_y, y_column)
    delta = 100
    epoch_number = 0
    while tolerance < delta:
        predicted_y = [linear_regression_function(beta0, beta1, x) for x in x_column]
        difference_y_values = [(expected - predicted) for (expected, predicted) in zip(y_column, predicted_y)]
        derivative_beta0 = - (2 / len(difference_y_values)) * np.sum(difference_y_values)
        sum_value = sum(x * y for x, y in zip(x_column, difference_y_values))
        derivative_beta1 = - (2 / len(difference_y_values)) * sum_value
        beta0 = beta0 - lr * derivative_beta0
        beta1 = beta1 - lr * derivative_beta1
        predicted_y = [linear_regression_function(beta0, beta1, x) for x in x_column]
        current_loss = loss_function(predicted_y, y_column)
        delta = loss - current_loss
        loss = current_loss
        print("Epoch %s: beta0 = %s; beta1 = %s; Loss function value = %s;" % (epoch_number, beta0[0], beta1[0], current_loss[0]))
        epoch_number += 1

    return {
        "Iterations"         : epoch_number,
        "beta0"              : beta0[0],
        "beta1"              : beta1[0],
        "Loss function value": loss[0]
    }

results = train(train_data)
results
```

```python
test_data = get_data("lab_1_test.csv")
plot_data(test_data, "Test data", 'bo')
```

```python
x_train, y_train = list(train_data[:, 1]), list(train_data[:, 2])
x_test, y_test = list(test_data[:, 1]), list(test_data[:, 2])
plt.plot(x_train, y_train, 'yo',
         x_test, y_test, 'bo',
         [0, 1], [linear_regression_function(results["beta0"], results["beta1"], 0), linear_regression_function(results["beta0"], results["beta1"], 1)])
```

```python
predicted_y = [linear_regression_function(results["beta0"], results["beta1"], x) for x in x_test]
print("Loss value: %s" % loss_function(predicted_y, y_test))
```
